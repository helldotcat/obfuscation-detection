```python
repos_files = [
    os.path.join(path, name) 
    for path, _, files in os.walk(DATA_REPOS)
    for name in files
    if name.endswith('.js') and name.count('.') == 1
]

def process_file(file):
    src = file
    dst = os.path.join(DATA_JS_FILES, file.lstrip(DATA_REPOS).replace('/', '_'))

    if os.path.exists(src):
        shutil.copy(src, dst)


with multiprocessing.Pool(processes=os.cpu_count() - 1) as pool:
    for _ in tqdm.tqdm(pool.imap_unordered(process_file, repos_files)):
        pass


print('JS files in repositories: ', len([_ for path, _, files in os.walk(DATA_JS_FILES) for name in files]))
```

```python
def process_file(file):
    input_file = file
    output_file = os.path.join(DATA_FUNCS, file.lstrip(DATA_JS_FILES))

    command = [
        'node',
        'scripts/parse_functions.js',
        input_file,
        output_file
    ]

    subprocess.run(command)


with multiprocessing.Pool(processes=os.cpu_count() - 1) as pool:
    for _ in tqdm.tqdm(pool.imap_unordered(process_file, files_to_funcs_sample), 
                       total=len(files_to_funcs_sample)):
        pass
```

```python
import os

STUNNIX_OUTPUT_FILES = 'data.eval/obfuscated/stunnix'
JO_FREE_OUTPUT_FILES = 'data.eval/obfuscated/jo_free_bak'
JO_BASIC_OUTPUT_FILES = 'data.eval/obfuscated/jo_basic'

stunnix_samples = sorted(
    [os.path.join(path, name) for path, _, files in os.walk(STUNNIX_OUTPUT_FILES) for name in files]
)

jo_free_samples = sorted(
    [os.path.join(path, name) for path, _, files in os.walk(JO_FREE_OUTPUT_FILES) for name in files]
)

jo_basic_samples = sorted(
    [os.path.join(path, name) for path, _, files in os.walk(JO_BASIC_OUTPUT_FILES) for name in files]
)


print(f'Stunnix: {len(stunnix_samples)}\nJO free: {len(jo_free_samples)}\nJO basic: {len(jo_basic_samples)}')
```

```python
import json


with open('data.files/feature_columns.json') as f:
    FEATURE_COLUMNS = json.load(f) 
```

```python
import sys
sys.path.extend([os.path.dirname(os.getcwd())])

from converter import converter


def process_file(code):
    try:
        features_extractor = converter.AstToCsvConverter()
        features_extractor.parse_from_string(code=code)
        features_extractor.convert()
        features_dict = features_extractor.get_features()
    except Exception as e:
        print(e)
        return
        
    feature_vector = []

    nodes_count = sum(features_dict['nodes'].values())
    nodes_pairs_count = sum(features_dict['node_pairs'].values())

    for feature in FEATURE_COLUMNS:
        if feature == 'v_types':
            feature_vector.append(
                len(features_dict['nodes'])
            )
            continue

        if feature == 'e_types':
            feature_vector.append(
                len(features_dict['node_pairs'])
            )
            continue
        
        key_first, key_second = feature.split(';')

        if key_first == 'nodes':
            feature_vector.append(
                features_dict[key_first].get(key_second, 0) / nodes_count
            )
        elif key_first == 'node_pairs':
            feature_vector.append(
                features_dict[key_first].get(key_second, 0) / nodes_pairs_count
            )
        else:
            feature_vector.append(
                features_dict[key_first].get(key_second, 0)
            )
            
    return feature_vector
```

```python
from catboost import CatBoostClassifier
from sklearn import metrics


model = CatBoostClassifier()
model.load_model('data.files/detection_model.save')
```

```python
def run_test(name, samples):
    X_test = []

    for file in tqdm.tqdm(samples):
        try:
            with open(file, 'r') as f:
                code = f.read()
        except Exception as e:
            print(e)
            continue

        if name == 'jo_free':
            code = code[545:] # remove disclaimer in comments
        
        result = process_file(code)
        if result is not None:
            X_test.append(result)

    X_test = np.array(X_test)
    y_test = np.ones(X_test.shape[0])
    y_pred = model.predict(X_test)
    
    print(f'!!!!! {name}')
    print(metrics.classification_report(y_test, y_pred, digits=4))
```

```python
import random
import tqdm
import numpy as np

jo_free_samples = sorted(jo_free_samples)

random.seed(144)
random.shuffle(jo_free_samples)

run_test('jo_free', jo_free_samples[:501])
```

```python
stunnix_samples = sorted(stunnix_samples)

random.seed(144)
random.shuffle(stunnix_samples)

run_test('stunnix', stunnix_samples[:501])
```

```python
jo_basic_samples = sorted(jo_basic_samples)

random.seed(144)
random.shuffle(jo_basic_samples)

run_test('jo basic', jo_basic_samples[:501])
```
